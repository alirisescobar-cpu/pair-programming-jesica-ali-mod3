Crea un array 2D de forma (4, 6) con números aleatorios decimales (no deben estar comprendidos solo entre 0 y 1). Accede a la última fila de la matriz. Los decimales deben estar redondeados a 2 decimales.

```python
import numpy as np

# Creamos 10 números entre 0 y 5
array_decimales = np.random.uniform(1, 24, 20)

# Redondeamos a 2 decimales
array_final = np.round(array_decimales, 2)

print(array_final)
```

```python
# Creamos la matriz 2D
# low=10, high=50, size=(filas, columnas)
matriz_2d = np.random.uniform(1, 9, size=(4, 6))

# Redondeamos a 2 decimales
matriz_final = np.round(matriz_2d, 3)

print("Mi matriz de 4x6 con decimales:")
print(matriz_final)
```

```python
redondeoultima = np.round(matriz_final[-1], 1)
```

```python
redondeoultima
```

Crea un array de tres 3D con elementos entre el 0 y 1 de forma (2, 3, 5) cuyos valores solo tengan 1 decimal
