```python
#Usando el iloc extrae todas las columnas de los países que estén en posición impar. 
#La primera que nos tendría que salir es Japón y la última Corea del Norte.

###RECORDAMOS  dataframe.loc[[fila(S)], [columna(s)]]

paises_posicion_impar=medallas.iloc[1::2, :]

#1::2 → selecciona filas impares

#: → selecciona todas las columnas

#Resultado → un DataFrame con todas las columnas, pero solo de las filas en posiciones impares

paises_posicion_impar
```

```python
#Utilizando el código del ejercicio anterior, selecciona unicamente las columnas , "Plata" y "Bronce".
paises_posicion_impar_plata_bronce=medallas.iloc[1::2, [1,2]]

#¿porque una cosa va sin corchetes y la otra con corchetes?

####tiene que ver con la diferencia entre slices y listas en Python/pandas

#1.Antes de la coma: filas (1::2) ----->

#1::2 → es un slice, no una lista
#Los slices en Python se escriben sin corchetes, con formato start:stop:step
#Selecciona filas en posiciones impares: 1, 3, 5 …

#✅ No necesitamos corchetes extra porque ya es un slice válido.

#2.Después de la coma: columnas ([1, 2]) ---->
#[1, 2] → lista de posiciones

#Necesitamos corchetes porque queremos seleccionar columnas específicas

#Aquí no usamos slice (start:stop:step), sino elementos exactos: columna 1 y columna 2



paises_posicion_impar_plata_bronce
```

```python
#Contesta a las siguientes preguntas:

#¿Qué país consiguió más de 200 medallas de oro? 
condicion=medallas["Oro"]>200 #---> medallas["Oro"] --->selecciona la columna de Oro,> 200 → devuelve un Boolean Series con True para los que cumplen la condición
paises_mas_200_oro = medallas[condicion]
print(paises_mas_200_oro)


###ALTERNATIVA EN UNA SOLA LINEA

###El primer medallas (antes de los corchetes):Este es el DataFrame completo
###El segundo medallas (dentro de los corchetes):Aquí estamos accediendo a la columna "Oro" del DataFrame medallas
####### medallas["Oro"] → devuelve una Serie con los valores de oro

paises_mas_200_oro = medallas[medallas["Oro"] > 200]
paises_mas_200_oro
```

```python
#¿Qué país consiguió más de 200 medallas de oro y más de 1000 de plata?

Condicion2=(medallas["Oro"]>200) & (medallas["Plata"]>1000)

oro200_plata1000=medallas[Condicion2]

oro200_plata1000
```

```python
#¿Qué país consiguió más de 200 medallas de oro o más de 700 de bronce?

condicion3=(medallas["Oro"]>200) | (medallas["Bronce"]>700)
oro200_bronce700=medallas[condicion3]

oro200_bronce700
```

Paso 2: Sumar las columnas fila por fila

-En pandas, podemos usar la función .sum() con el parámetro axis=1:
- axis=0 → suma columnas verticalmente (por cada columna)

- axis=1 → suma horizontalmente, fila por fila

```python
Total_medallas= medallas[["Oro", "Plata", "Bronce"]].sum(axis=1)
medallas.insert(3, "Total", Total_medallas)

#4 → posición de la columna (en este ejemplo, después de Oro, Plata y Bronce)

# "Total" → nombre de la columna

#Total_medallas → la Serie con la suma de medallas por país

```

Diferencia entre insert y asignación directa

medallas["Total"] = ... → añade la columna al final del DataFrame

medallas.insert(pos, "Total", ...) → añade la columna en la posición que quieras

💡 Tip:

Siempre que uses insert, la posición debe estar entre 0 y el número actual de columnas.

Si intentas insertar en una posición fuera de rango, pandas lanzará un error.
