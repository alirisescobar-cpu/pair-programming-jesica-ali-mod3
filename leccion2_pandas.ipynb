
1.¿Qué diferencia hay entre una Serie y un DataFrame?
-La serie es unidimensional, el dataframe es bidimensional.
-La serie admite solo un tipo de dato y el dataframe distintos

```python
import numpy as np
import pandas as pd
```

```python
#2.Es el turno de los DataFrames:
##Crea un array de dos dimensiones con 4 filas y 3 columnas.

array_paises= np.empty((4, 3))
print(array_paises)
```

```python
#Primero lo hemos convertido en un DataFrame
#Convierte el array en un DataFrame donde las columnas se llamen "España", "Francia" e "Italia"
df_paises=pd.DataFrame(array_paises)
df_paises
```

```python
#Descarga el archivo "medallas" y carga el DataFrame.
medallas=pd.read_csv("files/medallas.csv",index_col="País")
```

```python
#Usando el loc:Usando el loc extrae todas las columnas de la fila de Tailandia.
tailandia = medallas.loc[medallas['País'] == 'Tailandia', :]
```

```python
#Usando el loc extrae todas las columnas de China e Irán.

china_iran = medallas.loc[['China', 'Irán'], :]
china_iran
```
