```python
lista_paises = ["population", "density", "life_expectancy", "fertility_rate", "gdp"]
```

```python
def resumen_estadistica1(df):
    for columna in lista_paises:
        serie=df[columna]

        print(f"La media es {round(df[columna].mean(), 2)}")
        print(f"La mediana es {round(df[columna].median(), 2)}")
        print(f"La desviación estándar es {round(df[columna].std(), 2)}")
        print(f"El valor mínimo es {round(df[columna].min(), 2)}")
        print(f"El valor máximo es {round(df[columna].max(), 2)}")
```

```python
def resumen_estadistica(df, lista_columnas):
    # Iteramos por cada nombre de columna que hay en la lista que le pasamos
    for col in lista_columnas:
        # Extraemos la columna una sola vez para trabajar con ella (Series de Pandas)
        serie = df[col]
        
        print(f"--- Análisis de la columna: {col.upper()} ---")
        print(f"La media es: {round(serie.mean(), 2)}")
        print(f"La mediana es: {round(serie.median(), 2)}")
        print(f"La desviación estándar es: {round(serie.std(), 2)}")
        print(f"El valor mínimo es: {round(serie.min(), 2)}")
        print(f"El valor máximo es: {round(serie.max(), 2)}")
        print("-" * 40) # Una línea separadora para que se lea mejor en consola
```

```python
resumen_estadistica(df, lista_paises)
```
